# [dbo].[sp_ObtenerListado]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ObtenerListado]    Script Date: 16/01/2023 05:03:59 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create proc [dbo].[sp_ObtenerListado](
@ignorar_primeros int = 0,
@cantidad_filas int = 10,
@filtro varchar(100)
)as
begin 
		select IdBaleros, Marca, Codigo, Precio from Baleros
		where IdBaleros like '%' + @filtro + '%' or Marca like '%' + @filtro + '%' or Codigo like '%' + @filtro + '%'
		order by IdBaleros
		offset @ignorar_primeros rows
		fetch next @cantidad_filas rows only;
end
GO
```
